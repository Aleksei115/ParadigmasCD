````r
sumar_columnas <- function(matriz){
    ans <- list()
    for i in 1:ncol(matriz){
        suma <- 0
        for j in 1:nrow(matriz){
            suma <- suma + matriz[i][j]
        }
        ans.
    }
}


matriz_ejemplo <- matrix(1:20,4,5)

resultados <- sumar_columnas(matriz_ejemplo)

print(resultados)
````

Cambiar el marcador con la propiedad `pch`

Cambiamos el punto por la X

````r
plot(c(1,2,3,4),c(4,3,2,1),pch='x')
````

Agregamos etiquetas a los ejes con `xlab` y `ylab`

````r
plot(c(1,2,3,4),c(4,3,2,1),pch='x',xlab = 'axisX',ylab = 'axisY')
````

Una vez creada la gráfica se pueden agregar diferentes elementos como

*   `points(c(1,4),c(1,5))`
*   `lines(c(1,4),c(1,5),col='red')`






````r
plot(c(1,2,3,4,5),c(5,4,3,2,1),pch='x',xlab = 'axisX',ylab = 'axisY')
points(c(1,4),c(1,5))
lines(c(1,4),c(1,5),col='red')
````
